```python
from nba_api.stats.static import teams
nba_teams = teams.get_teams()

magic = [team for team in nba_teams if team['abbreviation'] == 'ORL'][0]
magic_id = magic['id']
print(f'magic id: {magic_id}')
```

```python

from nba_api.stats.library.parameters import Season
from nba_api.stats.library.parameters import SeasonType
from time import sleep

from nba_api.stats.endpoints import leaguegamefinder

# Query for games where the Celtics were playing
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=magic_id)
# The first DataFrame of those returned is what we want.
games = gamefinder.get_data_frames()[0]
```

```python
dfs = []
from nba_api.stats.endpoints import playbyplayv3
import time
count = 0

for game_id in games["GAME_ID"]:
    count += 1
    print(f'Getting play by play for game {count}')
    dfs.append(
        playbyplayv3.PlayByPlayV3(game_id=game_id)
        .get_data_frames()[0]
    )
    time.sleep(0.2)
```

```python
all_games = []
seasons = ['2019', '2020']
for season in seasons:
    print(f"Fetching {season}")
    
    gamefinder = leaguegamefinder.LeagueGameFinder(
        season_nullable=season,
        player_or_team_abbreviation="ORL",
        league_id_nullable="00"
    )
    
    df = gamefinder.get_data_frames()[0]
    all_games.append(df)
    time.sleep(1)
```
