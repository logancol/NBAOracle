```python
from nba_api.stats.endpoints import leaguegamefinder
import pandas as pd
import time
from nba_api.stats.static import teams
dfs = []
season_ids = [
 '1996-97', '1997-98', '1998-99', '1999-00',
 '2000-01', '2001-02', '2002-03', '2003-04',
 '2004-05', '2005-06', '2006-07', '2007-08',
 '2008-09', '2009-10', '2010-11', '2011-12',
 '2012-13', '2013-14', '2014-15', '2015-16',
 '2016-17', '2017-18', '2018-19', '2019-20',
 '2020-21', '2021-22', '2022-23', '2023-24',
 '2024-25', '2025-26'
]
for season_id in season_ids:
    df = leaguegamefinder.LeagueGameFinder(season_nullable=season_id, season_type_nullable="Regular Season").get_data_frames()[0]
    print('getting games from season')
    dfs.append(df)
    time.sleep(.8)

result = pd.concat(dfs)
```

```python
abbrevs = result['TEAM_ABBREVIATION'].unique()
abrev_id_map = {}
ids = []
for abrev in abbrevs:
    id = result[result['TEAM_ABBREVIATION'] == abrev].iloc[0]['TEAM_ID']
    abrev_id_map[abrev] = int(id)
```
